```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/IT Salary Survey EU 2018.csv")
df = pd.read_csv("/T Salary Survey EU 2019.csv")
df = pd.read_csv("/IT Salary Survey EU  2020.csv")

```

```python
sns.histplot(data=df, x='Age')
plt.show()

sns.histplot(data=df, x='Timestamp')
plt.show()
```
